## Handling Missing Values 

**It seems like there are quite a few null values in need of treatment. It is important to understand whether the null values are null because data were not collected or because there is some meaning into it. I will check each variable seperately in order to make sense. Perhaps, the area will give some insight**.

```python
#Let's start with Savanna fires. 
#I will make a new df series  'bool_series', that holds the boolean values of whether the Savanna fires column contains null values.
df['bool_series'] = pd.isnull(df["Savanna fires"])
```

```python
df.loc[df['bool_series'], 'Area'].unique()
```

**It seems like Holy See is the only area that has no records of Savanna values. This makes sence since it is not possible that such fire events occur in the city of Vatican. I will also see what other null values this Area has. It is a special case of an area since many features such as Forest fires, Savanna fires, manure, and tropical forest fires or farms do not exist**.

```python
null = df[df['Area'] == 'Holy See'].isnull()
```

```python
null.sum()
```

**As expected, those features contain null values because in reality they are non existent in the Holy see and hence, all the null values will be filled with 0**

```python
df.loc[df['Area'] == 'Holy See'] = df[df['Area'] == 'Holy See'].fillna(0)
```

**Lets Check Forest Fires**

```python
df['bool_series'] = pd.isnull(df["Forest fires"])
```

```python
df.loc[df['bool_series'], 'Area'].unique()
```

**Only Monaco and San Marino have null Forest fires values. Thats also reasonable because there are no forest in those two areas**

```python
df.loc[(df['Area'] == 'Monaco')|(df['Area'] == 'San Marino'), ['Forest fires']] = df.loc[(df['Area'] == 'Monaco') | (df['Area']=='San Marino'), ['Forest fires']].fillna(0)
```

**Let's proceed now with Fires in humid tropical forests**

```python
df['bool_series'] = pd.isnull(df['Fires in humid tropical forests'])
```

```python
df['bool_series'].sum()
```

```python
df.loc[df['bool_series'], 'Area'].unique()
```

**Those 4 areas do not have humid tropical forest and hence the null values will be replaced to 0**

```python
df.loc[df['bool_series'], 'Fires in humid tropical forests'] = df.loc[df['bool_series'], 'Fires in humid tropical forests'].fillna(0)
```

**Now all the missing values that can be filled reasonably(MNAR - Missing Data Not at Random) have been corrected. The rest of the variables include many missing values, most probably because there are simply no such records present in the dataset at random (MAR - Missing Data at Random). Hence, those values will be filled using linear regression under the MICE (Multiple Imputation by Chained Equation) framework**

```python
num_features = [col for col in df.columns]
```

```python
num_features.remove('Area')
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
lr = LinearRegression()
imp = IterativeImputer(estimator=lr,missing_values=np.nan,  max_iter=10, verbose=2, imputation_order='roman',random_state=0)
X=imp.fit_transform(df.iloc[:, 1:])
```

```python
df2 = pd.DataFrame(X, columns = num_features)
```

```python
#  Sort for correct YoY calculations 
df = df.sort_values(['Area', 'Year']).reset_index(drop=True)
```

```python
#  Year-over-Year emission change ─────────────────────────────────────
df['Prev_Year_Emission'] = df.groupby('Area')['total_emission'].shift(1)
df['YoY_Emission_Change_kt'] = (df['total_emission'] - df['Prev_Year_Emission']).round(2)
df['YoY_Emission_Growth_Pct'] = (
    (df['YoY_Emission_Change_kt'] / df['Prev_Year_Emission'].replace(0, np.nan)) * 100
).round(2)
```

```python
# Year-over-Year temperature change 
df['Prev_Year_Temp'] = df.groupby('Area')['Average Temperature °C'].shift(1)
df['YoY_Temp_Change'] = (df['Average Temperature °C'] - df['Prev_Year_Temp']).round(3)
```

```python
#  Decade column 
df['Decade'] = (df['Year'] // 10 * 10).astype(str) + 's'
```

```python
# Population & urbanization 
df['Total_Population'] = df['Total Population - Male'] + df['Total Population - Female']
df['Emission_Per_Capita'] = (
    df['total_emission'] / df['Total_Population'].replace(0, np.nan)
).round(4)
df['Urbanization_Rate_Pct'] = (
    (df['Urban population'] / df['Total_Population'].replace(0, np.nan)) * 100
).round(2)
```

```python
#  Sector aggregations 
df['Land_Fire_Emissions'] = df[[
    'Savanna fires', 'Forest fires',
    'Fires in organic soils', 'Fires in humid tropical forests'
]].sum(axis=1).round(2)

df['Agriculture_Emissions'] = df[[
    'Crop Residues', 'Rice Cultivation', 'Manure applied to Soils',
    'Manure left on Pasture', 'Manure Management',
    'On-farm energy use', 'On-farm Electricity Use'
]].sum(axis=1).round(2)

df['Industrial_Emissions'] = df[[
    'IPPU', 'Pesticides Manufacturing', 'Fertilizers Manufacturing'
]].sum(axis=1).round(2)

df['Food_System_Emissions'] = df[[
    'Food Transport', 'Food Household Consumption', 'Food Retail',
    'Food Packaging', 'Food Processing', 'Agrifood Systems Waste Disposal'
]].sum(axis=1).round(2)
```

```python
# Emission category bins 
def emission_category(val):
    if val < 5000:      return 'Low (<5K kt)'
    elif val < 20000:   return 'Medium (5K-20K kt)'
    elif val < 100000:  return 'High (20K-100K kt)'
    else:               return 'Very High (>100K kt)'

df['Emission_Category'] = df['total_emission'].apply(emission_category)
```

```python
# Temperature bands 
def temp_band(t):
    if t < 0.5:    return 'Low (<0.5C)'
    elif t < 1.0:  return 'Moderate (0.5-1C)'
    elif t < 1.5:  return 'Elevated (1-1.5C)'
    else:          return 'Critical (>1.5C)'

df['Temp_Band'] = df['Average Temperature °C'].apply(temp_band)
```

```python
#  Cumulative emission per country 
df['Cumulative_Emission_kt'] = df.groupby('Area')['total_emission'].cumsum().round(2)
```

```python
#  Dominant sector per row 
sector_cols = [
    'Land_Fire_Emissions', 'Agriculture_Emissions',
    'Industrial_Emissions', 'Food_System_Emissions'
]
df['Dominant_Sector'] = (
    df[sector_cols]
    .idxmax(axis=1)
    .str.replace('_Emissions', '')
    .str.replace('_', ' ')
)
```

```python
# ── Step 12: Drop helper columns ──────────────────────────────────────────────
df.drop(columns=['Prev_Year_Emission', 'Prev_Year_Temp'], inplace=True)
```

```python
#  Reorder columns (key columns first) 
id_cols = ['Area', 'Region', 'Year', 'Decade']

calc_cols = [
    'total_emission', 'YoY_Emission_Change_kt', 'YoY_Emission_Growth_Pct',
    'Emission_Category', 'Emission_Per_Capita', 'Cumulative_Emission_kt',
    'Land_Fire_Emissions', 'Agriculture_Emissions',
    'Industrial_Emissions', 'Food_System_Emissions', 'Dominant_Sector',
    'Average Temperature °C', 'YoY_Temp_Change', 'Temp_Band',
    'Total_Population', 'Urbanization_Rate_Pct'
]

pop_cols = [
    'Total Population - Male', 'Total Population - Female',
    'Rural population', 'Urban population'
]

raw_cols = [
    c for c in df.columns
    if c not in id_cols + calc_cols + pop_cols
]

df = df[id_cols + calc_cols + raw_cols + pop_cols]
```
